```python
%matplotlib inline
import pandas as pd
import tushare as ts
```

```python
data = ts.get_hist_data('510050','2018-01-01')
data = data.sort_index()
data
```

```python
df = pd.DataFrame()
df['close'] = data['close']
df['change'] = data['close']- data['close'].shift(1)
df
```

```python
df['ma5'] = df['close'].rolling(window = 5,center = False).mean()
df['ma20'] = df['close'].rolling(window = 20,center = False).mean()
df
```

```python
df = df.dropna()
df
```

```python
df['pos'] = 0
df['pos'][df['ma5'].shift(1)>df['ma20'].shift(1)] = 10000
df['pos'][df['ma5'].shift(1)<=df['ma20'].shift(1)] = -10000
df
```

```python
df['fee'] = 0
df['fee'][df['pos'] != df['pos'].shift(1)] = df['close']*0.0003*20000
df['fee'][0] = 0
df
```

```python
df['pnl'] = df['change']*df['pos']
df['netpnl'] = df['pnl'] - df['fee']
df['cumpnl'] = df['netpnl'].cumsum()
df
```

```python
df['cumpnl'].plot()
```
